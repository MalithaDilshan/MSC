```python
x = np.arange(-1*np.pi, np.pi, 0.1)
y = np.sin(x)
plt.plot(x,y,'ro-.')
plt.title('y = Sin(x)', {'fontsize':20})
plt.xlabel('Angle (rad)')
plt.ylabel('Sin(x)')
plt.grid()
plt.legend('sin(x)')
plt.show()
```

```python
# Plot multiple data sets
x = np.arange(-2*np.pi, 2*np.pi, 0.01)
y1 = np.sin(x)
y2 = np.sin(x - np.pi/2)
y3 = np.sin(x - np.pi)

#Method 01

plt.plot(x,y1, x,y2, x,y3)
plt.title('y = Sin(x)', {'fontsize':20})
plt.xlabel('Angle (rad)')
#plt.ylabel('Sin(x)')
plt.grid()
plt.legend(['sin(x)', 'sin(x-pi/2)', 'sin(x - pi)'])
plt.show()
```

```python
# Plot multiple data sets
x = np.arange(-2*np.pi, 2*np.pi, 0.01)
y1 = np.sin(x)
y2 = np.sin(x - np.pi/2)
y3 = np.sin(x - np.pi)

#Method 02

plt.plot(x,y1, linewidth=3)
plt.plot(x,y2)
plt.plot(x,y3)
plt.grid()
plt.legend(['sin(x)', 'sin(x-pi/2)', 'sin(x-pi)'])
plt.show()
```

```python
#Sub plots

x = np.arange(-2*np.pi, 2*np.pi, 0.01)
y1 = np.sin(x)
y2 = np.sin(x - np.pi/2)
y3 = np.sin(x - np.pi)

fig, (ax1,ax2,ax3) = plt.subplots( 1,3)
ax1.plot(x,y1)
ax2.plot(x,y2)
ax3.plot(x,y3)
fig.tight_layout()
fig.show()
```

```python
x = np.array([10,20,30,40,50])
y = np.array([20,30,40,50,70])

fig, ax = plt.subplots()
ax.plot(x,y)
fig.show()
```

```python
#Bar graphs

x = np.array([0,1,2,3,4])
y = np.array([20,30,40,50,60])

fig, ax = plt.subplots()
ax.bar(x,y)
ax.set_title('Bar')
ax.set_xlabel( 'X')
ax.set_ylabel('Y')
fig.show()
```

```python
x = np.arange(5)
y = np.array([20,30,40,50,60])
z = np.array([20,30,45,53,47])

fig, ax = plt.subplots()
ax.bar(x-0.2,y, width=0.4)
ax.bar(x+0.2,z, width=0.4)
ax.set_title('Bar')
ax.set_xlabel('X')
ax.set_ylabel('Y')
fig.show()
```

```python
# scatter plots
np.random.seed(1)
N = 20
x = np.random.rand(N)
y = np.random.rand(N)
fig, ax = plt.subplots()
ax.scatter(x,y)
fig.show()
```

```python
z = np.hstack((np.zeros(10), np.ones(10)))
fig, ax = plt.subplots()
ax.scatter(x, y, c=z)
fig.show()
```

```python
#pie

x = [5,15,20,25]
l = ['A','B','C','D']
fig, ax = plt.subplots()
ax.pie(x, labels=l, autopct='%1.1f%%')
fig.show()
```

```python
# Histograms
mu, sigma = 0, 0.1
x = np.random.normal(mu,sigma,1000)
fig, ax = plt.subplots()
ax.hist(x,100)
fig.show()
```
