```python
# Session 6 - Machine Learning
from sklearn import datasets
iris = datasets.load_iris()
print(iris)
```

```python
#Assign data and target ( class label )

X = iris.data
Y = iris.target

#Print first 10 rows of X and Y
print(X[:10])
print(Y[:10])
print(X.size)
print(X[-10:])
```

```python
import matplotlib.pyplot as plt
plt.plot(X[:,2], X[:,0], 'ro')
plt.show()
```

```python
plt.plot(X[Y[:]==0,2], X[Y[:]==0,0], 'ro')
plt.plot(X[Y[:]==1,2], X[Y[:]==1,0], 'go')
plt.plot(X[Y[:]==2,2], X[Y[:]==2,0], 'bo')
plt.title('Iris dataset', {'fontsize':20})
plt.xlabel('Petal.Length')
plt.ylabel('Sepal.Length')
plt.legend(['Setosa', 'Versicolor', 'Virginica'])
plt.show()
```

```python
# Split data as training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.33, random_state=1)
print( X_train.shape)
print( X_test.shape)
print( y_train.shape)
print( y_test.shape)
```

```python
# Build decision tree modal
from sklearn import tree
classifier = tree.DecisionTreeClassifier(random_state=1)
clf = classifier.fit(X_train, y_train)
```

```python
tree.plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, rounded=True, proportion=True)
plt.show()
predictions = classifier.predict(X_test)
print(predictions)

```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import tree
import matplotlib.pyplot as plt

# Load the dataset
bc = datasets.load_breast_cancer()
X = bc.data
Y = bc.target

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=1)

# Print shapes
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)

# Train the decision tree classifier
classifier = tree.DecisionTreeClassifier(random_state=1)
clf = classifier.fit(X_train, y_train)

# Increase the plot size
plt.figure(figsize=(30, 20))  # Set the figure size (width, height)

# Plot the decision tree
tree.plot_tree(
    clf,
    feature_names=bc.feature_names,
    class_names=bc.target_names,
    filled=True,
    rounded=True,
    proportion=True,
)

# Show the plot
plt.show()
```

```python
# make prediction
predictions = classifier.predict(X_test)
print(predictions)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predictions))
```

```python
# Confusion Metrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, predictions))
```

```python
from sklearn.metrics import classification_report
target_name = ['Yes','No']
print(classification_report(y_test,predictions, target_names=target_name))
```

```python
#final modal
clf = classifier.fit(X, Y)
tree.plot_tree(
    clf,
    feature_names=bc.feature_names,
    class_names=bc.target_names,
    filled=True,
    rounded=True,
    proportion=True,
)

# Show the plot
plt.show()
```
